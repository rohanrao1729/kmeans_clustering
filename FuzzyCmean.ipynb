```python
import math
def euclideanDist(x,y):
    sum=0
    leng=len(x)
    for i in range(0,leng):
        sum+=(x[i]-y[i])*(x[i]-y[i])
    sum=math.sqrt(sum)
    return sum 
```

```python
def initialiseCMeans(items,k):
    means=[]
    for i in range(0,k):
        means.append(items[i])
    return means
```

```python
print(initialiseCMeans([[1,2,3],[3,4,5],[5,5,7],[3,45,6]],2))
```

```python
def calcMembership(obj,cluster,means):
    ans=0
    clusInd=cluster
    normDist=euclideanDist(obj,means[clusInd-1])
    for i in range(len(means)):
        val=euclideanDist(obj,means[i])
        if(val==0):
            if(normDist==0):
                return 1
            else :
                return 0
            
        formal=float(normDist/val)
        formal=formal*formal
        ans+=formal
    ans1=1/ans
    return ans1
```

```python
print(calcMembership([2],2,[[3],[11]]))
```

```python
def updateHelperA(membership,items,k,ind):
    ans=0
    for i in range(len(membership)):
        ans+=(membership[i][ind-1]*membership[i][ind-1])*items[i][k]
    return ans    

def updateHelperB(membership,i):
    ans=0
    for j in range(len(membership)):
        ans+=(membership[j][i-1]*membership[j][i-1])
    return ans    
def updateClustCMean(membership,items,i):
    sumDen=updateHelperB(membership,i)
    mean=[]
    for j in range(len(items[0])):
        val=updateHelperA(membership,items,j,i)
        tans=val/sumDen
        mean.append(tans)
    return mean
def updateAllMeans(membership,items):
    means=[]
    for i in range(len(membership[0])):
        means.append(updateClustCMean(membership,items,i+1))
    return means    
```

```python
def calcMembershipAll(items,means):
    membersh=[[0 for ie in range(len(means))] for je in range(len(items))]
    for i in range(len(items)):
        for j in range(len(means)):
            membersh[i][j]=calcMembership(items[i],j+1,means)
                        
    return membersh 

def checkMembership(membership):
    label=[]
    for i in range(len(membership)):
        maxi=membership[i][0]
        ind=0
        for j in range(len(membership[0])):
            if membership[i][j]>maxi:
                maxi=membership[i][j]
                ind=j
        label.append(ind)
    return label   
def fuzzyCMeans(k,items,noOfIter=10):
    means=initialiseCMeans(items,k)
    epsilon=0.001
    for i in range(0,noOfIter):
        preMeans=means
        membership=calcMembershipAll(items,means)
        means=updateAllMeans(membership,items)
        c=0
        for j in range(len(means)):
            if(euclideanDist(means[j],preMeans[j])<epsilon):
                c+=1
            if(c==len(means)):
                break
    answer=checkMembership(membership)
    return answer;
```

```python
print(fuzzyCMeans(2,[[1,2,3,4],[2,3,4,5],[3,4,5,6],[1,2,3,5]]))
```
